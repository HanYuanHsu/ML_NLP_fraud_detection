```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
feature_matrix = pd.read_csv("jobs_feature.csv")
feature_matrixindustry_Accounting
```

```python
feature_matrix[feature_matrix['industry_Accounting']==1]
```

```python
X = feature_matrix.drop(['fraudulent', 'company_profile_length', 'description_length', 'requirements_length', 'benefits_length'], axis=1)
y = feature_matrix['fraudulent']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=42)
```

```python
rfm = RandomForestClassifier(n_estimators = 200, random_state=9)
rfm.fit(X_train,y_train)
```

```python
y_pred_rfm = rfm.predict(X_test)
y_test = np.array(y_test)
print(len(y_test)) # number of test samples
wrong_index = np.where(y_pred_rfm != y_test)
wrong_index
```

```python
pred_prob = rfm.predict_proba(X_test)
print(pred_prob[wrong_index, :])
print(y_test[wrong_index])
```

```python
#We should look at sensitivity instead!!!!!!
confusion_matrix(y_true=y_test, y_pred=y_pred_rfm)
# the (i,j)-entry is equal to the number of observations known to be in group i and
# predicted to be in group j
# TPR = entry (2,2) / sum of bottom row
```
