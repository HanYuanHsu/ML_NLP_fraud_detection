```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import zero_one_loss
```

```python
# colnames: a list of column names
# variables returned are:
# best: the best colname with one feature removed, that is, the colname with the least significant feature removed
# error: the error (loss) corresponding to that best 
def backward_selection_helper(colnames, X_train, X_test, y_train, y_test):
    best = colnames
    error = np.Inf
    
    for i in range(len(colnames)): 
        # remove item i
        colnames1 = colnames.copy()
        colnames1.pop(i)
        
        ##### random forest
        rfm = RandomForestClassifier(n_estimators = 100, oob_score=True)
        rfm.fit(X_train[colnames1], y_train)
        y_pred_rfm = rfm.predict(X_test[colnames1])
        l = zero_one_loss(y_test, y_pred_rfm) 
            
        if (l < error):
            error = l
            best = colnames1
    
    return best, error

#
# variables returned are:
# best: the best subset of features
# error: the error (loss) corresponding to that best 
def backward_selection(X_train, X_test, y_train, y_test):
    best = list(X_train.columns)
    error = np.Inf
    current_best = list(X_train.columns)
    current_error = np.Inf
    
    while len(current_best) > 1:
        current_best, current_error = backward_selection_helper(current_best, X_train, X_test, y_train, y_test)
        if current_error < error:
            error = current_error
            best = current_best
            
    return best, error   
```

```python
feature_matrix = pd.read_csv('jobs_feature.csv')
feature_matrix.head()
```

```python
X = feature_matrix.drop("fraudulent", axis = 1)
y = feature_matrix.fraudulent
y = y.astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
rfm = RandomForestClassifier(n_estimators = 100, oob_score=True)
rfm.fit(X_train,y_train)
rfm.score(X_train,y_train)
```

```python
y_pred_rfm=rfm.predict(X_test)

zero_one_loss(y_test, y_pred_rfm)
```

```python
backward_selection(X_train, X_test, y_train, y_test) # too slow
```
