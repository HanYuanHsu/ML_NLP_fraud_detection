```python
# finds most frequent words in description column (ignoring all the stop words)
top_N = 20
df = pd.read_csv("job_training_data.csv",
                 usecols=['description','requirements'])
stopwords = nltk.corpus.stopwords.words('english')
RE_stopwords = r'\b(?:{})\b'.format('|'.join(stopwords))
words = (df.description
           .str.lower()
           .replace([r'\|', RE_stopwords], [' ', ''], regex=True)
           .str.cat(sep=' ')
           .split()
)
rslt = pd.DataFrame(collections.Counter(words).most_common(top_N),
                    columns=['Word', 'Frequency']).set_index('Word')
rslt
```
